## Feature Engineering

```python
def age(born, today): # Function to calculate age
    born = datetime.strptime(born, "%Y-%m-%d").date()
    today = datetime.strptime(today, "%Y-%m-%d").date()
    return today.year - born.year- ((today.month, today.day) < (born.month, born.day))
```

```python
df['Dogum_Tarihi'] = df['Dogum_Tarihi'].astype(str)
df['Ilac_Baslangic_Tarihi'] = df['Ilac_Baslangic_Tarihi'].astype(str)
df['Ilac_Bitis_Tarihi'] = df['Ilac_Bitis_Tarihi'].astype(str)

df['Yas'] = df.apply(lambda row: age(row['Dogum_Tarihi'], row['Ilac_Baslangic_Tarihi']), axis=1)
```

```python
df['Ilac_Baslangic_Tarihi'] = pd.to_datetime(df['Ilac_Baslangic_Tarihi'])
df['Yan_Etki_Bildirim_Tarihi'] = pd.to_datetime(df['Yan_Etki_Bildirim_Tarihi'])
df['Ilac_Yan_Etki_Suresi'] = (df['Yan_Etki_Bildirim_Tarihi'] - df['Ilac_Baslangic_Tarihi']).dt.days
```

Here, I considered the duration of the drug use as the Yan_Etki_Bildirim_Tarihi date because it is important for us to know when the side effect occurred. The medication's end date depends on many other variables (such as reaching the doctor, tests, and the decision to stop the medication), so I made this choice.

I filled in the missing values for columns like city, allergy, chronic illness, and blood type with the label 'Unknown' because I believe replacing them with any method might bias the model, unlike using averages for numerical data which is more neutral. For the remaining numerical and gender data, I think I can fill in the missing values more effectively by using them to help each other.

```python
# Function for binary encoding
def binary_encode(df, cols_to_cat=None):
    if cols_to_cat is not None:
        assert set(cols_to_cat) <= set(df.columns)
        df[cols_to_cat] = df[cols_to_cat].astype('object')
        X = df.select_dtypes('object')  
    else:
       
        X = df.select_dtypes('object')

    encoder = ce.binary.BinaryEncoder(return_df=True)
    encoded_X = encoder.fit_transform(X)

    
    encoded_X = encoded_X.loc[:, (encoded_X != 0).any(axis=0)]
    
    return encoded_X
```

```python
# Funtion for one-hot encoding
def onehot_encode_features(df):
    df_encode = pd.DataFrame()
    for attr in df.columns:
        p = df[attr].str.get_dummies(sep=", ")
        p.columns = list(map(lambda x: attr + "_" + x, p.columns))
        df_encode = pd.concat([df_encode,p],axis=1)
    return df_encode

df_encode = onehot_encode_features(df3)    
```

# Principal Component Analysis (PCA)

PCA is a dimensionality reduction technique for data sets with many features or dimensions. It uses linear algebra to determine the most important features of a dataset. After these features have been identified, you can use only the most important features, or those that explain the most variance, to train a machine learning model and improve the computational performance of the model without sacrificing accuracy. 

```python
scaler = StandardScaler()
data_rescaled = scaler.fit_transform(df_last)

pca = PCA().fit(data_rescaled)

plt.rcParams["figure.figsize"] = (8,5)

fig, ax = plt.subplots()
xi = np.arange(1, 94, step=1)
y = np.cumsum(pca.explained_variance_ratio_)

plt.ylim(0.0,1.1)
plt.plot(xi, y, marker='o', linestyle='-', color='black')

plt.xlabel('Number of Components')
plt.xticks(np.arange(1, 94, step= 3)) 
plt.ylabel('Cumulative variance (%)')
plt.title('The number of components needed to explain variance')

plt.axhline(y=0.90, color='grey', linestyle='--')
plt.text(1.1, 1, '90% cut-off threshold', color = 'black', fontsize=16)

ax.grid(axis='x')
plt.tight_layout()
plt.savefig('pcavisualize_1.png', dpi=300)
plt.show()
```

Since there are approximately 93 columns in the dataset, applying PCA to the final data may be a reasonable choice. This way, we have both reduced dimensionality and structured our dataset at a certain size by measuring variance. As we can understand from the cumulative variance graph above, if we set the variance at around 90%, we can limit our components to approximately 64.

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_last)
pca = PCA(n_components=64)
df_pca = pca.fit_transform(df_scaled)
```
